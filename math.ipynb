```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] # 显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 显示负号
```

```python
x = np.arange(-10, 10, 0.1)
y = []
for t in x:
    y_1 = 1 / (1 + math.exp(-t))
    y.append(y_1)
plt.plot(x, y, label="sigmoid")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sigma')
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
#sin & cos曲线
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label="sin")
plt.plot(x,y2,label="cos",linestyle = "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```

```python
x = np.arange(-6, 6, 0.1)
a1 = 2
a2 = 3
a3 = 5
y1 = np.sinc(x) * np.sinc(x/a1)
y2 = np.sinc(x) * np.sinc(x/a2)
y3 = np.sinc(x) * np.sinc(x/a3)
plt.plot(x,y1,label="a=2")
plt.plot(x,y2,label="a=3",linestyle="--")
plt.title('sinc')
plt.legend()
plt.show()
```

```python
x = np.arange(0.1, 100, 0.1)        #从0.1到100，间隔0.1取点
y = []
for t in x:
    if t < 50:
        y_1 = 5000 / t
    else:
        y_1 = 200 - t * 2
    y.append(y_1)
plt.plot(x, y, label="qp-quality")
plt.xlabel("quality")
plt.ylabel("qp")
plt.title('qp')
plt.ylim(0, 500)
plt.legend()    #显示label
plt.show()
```

```python
x = np.arange(0.001, 0.05, 0.002)
y1 = np.log10(x)
y2 = np.log(x)
plt.plot(x, y1, label='log10(x)', marker='|', linestyle='dotted', color='purple')
plt.plot(x, y2, label='log(x)', linestyle='-.', color='yellow')
plt.xlabel('x')
plt.ylabel('y')
plt.title('log')
plt.legend()
plt.show()
```

```python
e = math.e
fig = plt.figure(figsize=(12,8), dpi=72)
x = np.arange(-10, 10, 0.01)
arsinh = np.log(x+np.sqrt(x**2+1))
sinh=0.5*(e**x-e**(-x))
cosh=0.5*(e**x+e**(-x))

plt.plot(x, sinh,label="双曲正弦")
plt.plot(x, arsinh,label="反双曲正弦")
plt.plot(x, cosh,label="双曲余弦")

plt.legend(loc='upper left')

plt.ylim(-10,10)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.show()
```

```python
x = np.arange(0.001, 5, 0.001)
y1 = np.log(x)
y2 = np.log2(x)
y3 = np.log10(x)

plt.plot(x, y1, label='log(x)')
plt.plot(x, y2, label='log2(x)')
plt.plot(x, y3, label='log10(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('log')
plt.legend()
plt.show()
```

```python
x1 = np.arange(-1, 1, 0.001)
y1 = np.arccos(x1)
x2 = np.arange(1, 5, 0.001)
y2 = np.arccosh(x2)

x3 = np.arange(-1, 1, 0.001)
y3 = np.arcsin(x3)
x4 = np.arange(-5, 5, 0.001)
y4 = np.arcsinh(x4)

plt.plot(x1, y1, label='acos(x)')
plt.plot(x2, y2, label='acosh(x)')
plt.plot(x3, y3, label='asin(x)')
plt.plot(x4, y4, label='asinh(x)')
plt.xlabel('x')
plt.ylabel('y')
# plt.title('cos')
plt.grid()
plt.legend()
plt.show()
```

```python
# 分段函数
x = np.arange(-10, 10, 0.001)

interval0 = [1 if (i < 0) else 0 for i in x]
interval1 = [1 if (0 <= i < 1) else 0 for i in x]
interval2 = [1 if (i >= 1) else 0 for i in x]

y = x * interval0 + np.cos(x) * interval1 + np.sin(x) * interval2

plt.plot(x, y)
plt.show()
```

```python
# 分段函数
x = np.linspace(-1, 5, 100)
y = np.piecewise(
    x,
    [x <= 0, (0 < x) & (x < 1), x >= 1],
    [lambda x: -x, lambda x: np.arcsin(x), lambda x: np.sin(x)]
)
plt.plot(x, y)
plt.show()
```

```python
# (min, max)取对数后均分为5段
pixel_min = 90
pixel_max = 730

y_min = np.log10(pixel_min)
y_max = np.log10(pixel_max)
delta = (y_max - y_min) / 4

arr = [y_min]
for i in range(3):
    arr.append(y_min + delta * (i + 1))
arr.append(y_max)
print(arr)

res = []
for i in range(len(arr)):
    res.append(np.floor(np.power(10, arr[i])))
print(res)
```

```python
#螺线
fig, ax = plt.subplots()
r = 10
n = 1000
theta = 0
for i in range(n):
    x = r * theta * math.cos(theta)
    y = r * theta * math.sin(theta)
    ax.plot(x, y, 'bo')
    theta += 0.1
plt.show()
```
